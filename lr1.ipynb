```python
import random
import copy
import math
import time

def matrix_create(check):
    if check == "Да" or check == "да":
        m = int(input("Введите количество строк: "))
        n = int(input("Введите количество столбцов: "))
        min_limit = int(input("Укажите начальный диапозон чисел для генерации: "))
        max_limit = int(input("Укажите конечный диапозон чисел для генерации: "))
        a = [[random.randint(min_limit, max_limit) for j in range(n)] for i in range(m)]
        return(a)
    elif check == "Нет" or check == "нет":
        m = 50
        n = 50
        min_limit = -250
        max_limit = 1000 + 14
        a = [[random.randint(min_limit, max_limit) for j in range(n)] for i in range(m)]
        return(a)

check = str(input("Хотите ввести числа? : "))
a = matrix_create(check)
print("Сгенерированная матрица")
for i in range(len(a)):
    print(a[i])

def selection(a):
    for row in range(len(a)):
        for col in range(len(a) - 1):
            for colS in range(col + 1, len(a)):
                if a[row][colS] < a[row][col]:
                    a[row][col], a[row][colS] = a[row][colS], a[row][col]
    return a

print("сортировка выбором")
select = copy.deepcopy(a)
start_time_select = time.time()
selecti = selection(select)
end_time_select = time.time()-start_time_select
for i in range(len(select)):
    print(selecti[i])
print("Время выполнения: ",'%.2f' %end_time_select,"sec")

def insertion_sort(a):
    for row in range(len(a)):
        for col in range (1,len(a[0])):
            key = a[row][col]
            j = col-1
            while j >=0 and key < a[row][j]:
                a[row][j+1] = a[row][j]
                j -= 1
            a[row][j+1] = key
    return a

print("сортировка вставкой")
select_insert = copy.deepcopy(a)
start_time_insert = time.time()
select_insert_1 = insertion_sort(select_insert)
end_time_insert = time.time()-start_time_insert
for i in range(len(select_insert)):
    print(select_insert_1[i])
print("Время выполнения: ",'%.2f' %end_time_insert,"sec")

def bubble_sort(a):
    for row in range(len(a)):
        for col in range(len(a[0])-1):
            for j in range(len(a[0])-col-1):
                if a[row][j] > a[row][j+1]:
                    a[row][j], a[row][j+1] = a[row][j+1], a[row][j]
    return a

print("Сортировка обменом(пузырьком)")
bubble = copy.deepcopy(a)
start_time_bubble = time.time()
bubble_1 = bubble_sort(bubble)
end_time_bubble = time.time()-start_time_bubble
for i in range(len(bubble)):
    print(bubble_1[i])
print("Время выполнения: ",'%.2f' %end_time_bubble,"sec")

def shellSort(a):
    n = len(a[0])
    k = int(2^n)
    interval = int(2**k -1)
    while interval > 0:
        for row in range(len(a)):
            for col in range(interval, n):
                temp = a[row][col]
                j = col
                while j >= interval and a[row][j - interval] > temp:
                    a[row][j] = a[row][j - interval]
                    j -= interval
                a[row][j] = temp
        k -= 1
        interval = int(2**k -1)
    return a

print("Сортировка Шелла")
shell = copy.deepcopy(a)
start_time_shell = time.time()
shell_1 = shellSort(shell)
end_time_shell = time.time()-start_time_shell
for i in range(len(shell)):
    print(shell_1[i])
print("Время выполнения: ",'%.2f' %end_time_shell,"sec")

def tournamentSort(a):
    tree = [None] * 2 * (len(a) + len(a) % 2)
    index = len(tree) - len(a) - len(a) % 2

    for i, v in enumerate(a):
        tree[index + i] = (i, v)

    for j in range(len(a)):
        n = len(a)
        index = len(tree) - len(a) - len(a) % 2
        while index > -1:
            n = (n + 1) // 2
            for i in range(n):
                i = max(index + i * 2, 1)
                if tree[i] != None and tree[i + 1] != None:
                    if tree[i][1] < tree[i + 1][1]:
                        tree[i // 2] = tree[i]
                    else:
                        tree[i // 2] = tree[i + 1]
                else:
                    tree[i // 2] = tree[i] if tree[i] != None else tree[i + 1]
            index -= n

        index, x = tree[0]
        a[j] = x
        tree[len(tree) - len(a) - len(a) % 2 + index] = None

def TournamentSort(a):
    for i in range(len(a)):
        tournamentSort(a[i])
    return a

print("Турнирная сортировка")
tour = copy.deepcopy(a)
start_time_tour = time.time()
tour_1 = TournamentSort(tour)
end_time_tour = time.time()-start_time_tour
for i in range(len(tour)):
    print(tour_1[i])
print("Время выполнения: ",'%.2f' %end_time_tour,"sec")


def heapify(a, n, i, row):
    largest = i
    l = 2 * i + 1
    r = 2 * i + 2

    if l < n and a[row][i] < a[row][l]:
        largest = l
    if r < n and a[row][largest] < a[row][r]:
        largest = r

    if largest != i:
        a[row][i], a[row][largest] = a[row][largest], a[row][i]
        heapify(a, n, largest, row)
def heapSort(a):
    for row in range(len(a)):
        n = len(a[0])
        for i in range(n//2, -1, -1):
            heapify(a, n, i, row)
        for i in range(n-1, 0, -1):
            a[row][i], a[row][0] = a[row][0], a[row][i]
            heapify(a, i, 0, row)
    return a

print("Пирамидальная сортировка")
heap = copy.deepcopy(a)
start_time_heap = time.time()
heap_1 = heapSort(heap)
end_time_heap = time.time()-start_time_heap
for i in range(len(heap)):
    print(heap_1[i])
print("Время выполнения: ",'%.2f' %end_time_heap,"sec")

def partition(a, low, high,row):
    pivot = a[row][(low + high) // 2]
    i = low - 1
    j = high + 1
    while True:
        i += 1
        while a[row][i] < pivot:
            i += 1

        j -= 1
        while a[row][j] > pivot:
            j -= 1

        if i >= j:
            return j

        a[row][i], a[row][j] = a[row][j], a[row][i]
def quick_sort(a):
    def _quick_sort(items, low, high,row):
        if low < high:
            split_index = partition(items, low, high,row)
            _quick_sort(items, low, split_index,row)
            _quick_sort(items, split_index + 1, high,row)

    for row in range(len(a)):
        _quick_sort(a, 0, len(a[row])-1,row)

    return a

print("Быстрая сортировка")
quick = copy.deepcopy(a)
start_time_quick = time.time()
quick_1 = quick_sort(quick)
end_time_quik = time.time()-start_time_quick
for i in range(len(quick)):
    print(quick_1[i])
print("Время выполнения: ",'%.2f' %end_time_quik,"sec")
```
